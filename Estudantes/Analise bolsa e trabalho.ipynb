# ConclusÃ£o

```python
print("A mÃ©dia de gastos geral Ã© de {}".format(n0))
print("A mÃ©dia de gastos para um bolsista que trabalha Ã© de {}".format(n1))
print("A mÃ©dia de gastos para um bolsista que nÃ£o trabalha Ã© de {}".format(n2))
print("A mÃ©dia de gastos para um nÃ£o bolsista que trabalha Ã© de {}".format(n3))
print("A mÃ©dia de gastos para um nÃ£o bolsista que nÃ£o trabalha Ã© de {}".format(n4))
```

```python
medias = data = [n0, n1, n2, n3, n4]
medias
```

```python
tabela_medias = pd.DataFrame(medias, 'geral bolsa_e_trabalho bolsa_sem_trabalho trabalho_sem_bolsa sem_bolsa_e_sem_trabalho'.split(), 
                             columns = ['MÃ©dias'])
tabela_medias.columns.name = 'Tipo'
tabela_medias
```

```python
tabela_medias = tabela_medias.reset_index()
tabela_medias
```

```python
tabela_medias.columns = ('Tipos', 'Media de Gastos')
tabela_medias
```

```python
tabela_medias = tabela_medias.sort_values(by = 'Media de Gastos', ascending = False)
```
