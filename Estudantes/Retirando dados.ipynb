# Removendo nulos

```python
import pandas as pd
```

```python
dados = pd.read_csv('dados/University-Students-Monthly-Expenses.csv')
dados
```

```python
dados.columns = ['Genero', 'Idade', 'Anos de Estudo', 'Moradia', 'Bolsa', 'Trabalho', 'Transporte', 'Fumante', 'Consome Alcool', 'Joga', 'Usa Cosmeticos', 'Assinaturas', 'Gastos Mensais']
```

```python
dados.head()
```

```python
dados.info()
```

```python
dados['Gastos Mensais'].info()
```

```python
selecao = dados["Gastos Mensais"].isnull()
dados[~selecao]
dados.dropna(subset = ['Gastos Mensais'], inplace = True)
dados.info()
```

```python
dados.to_csv('dados/tabela_geral_estudantes.csv', index = False)
```
