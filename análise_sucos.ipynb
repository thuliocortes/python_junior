```python
import pandas as pd
```

```python
dados = pd.read_csv('sucos.csv')
```

```python
dados.head(2)
```

```python
dados[{"Faturamento",'Unidade_Vendida', 'Custo_Variavel', 'Quantidade_Vendida'}].describe()
```

```python
import seaborn as sns
```

```python
sns.set_style("whitegrid")
sns.set_palette("CMRmap_r")
ax = sns.barplot(x = "Desc_Segmento", y = "Faturamento", data = dados)
ax.figure.set_size_inches(12,6)
ax.set_title('Faturamento por Segmento',fontsize = '18')
```

```python
ax = sns.barplot(x = "Desc_Estado", y = "Faturamento", data = dados)
ax.figure.set_size_inches(18,8)
ax.set_title('Faturamento por Estado',fontsize = '18')
```

```python
ax = sns.barplot(x = "Desc_Fabrica", y = "Faturamento", data = dados)
ax.figure.set_size_inches(12,6)
ax.set_title('Faturamento por FÃ¡brica',fontsize = '15')
```
