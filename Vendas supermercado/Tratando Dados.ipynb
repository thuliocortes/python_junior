```python
import pandas as pd 
```

```python
dados = pd.read_csv('supermarket_sales - Sheet1.csv')
dados
```

```python
dados = dados.drop(columns= ['Invoice ID', 'Branch', 'cogs'])
```

```python
dados.head()
```

```python
dados.columns
```

```python
renomear = {'City': 'cidade', 'Customer type': 'tipo_cliente', 'Gender': 'genero', 'Product line': 'tipo_produto', 
            'Unit price': 'preco', 'Quantity': 'quantidade', 'Tax 5%': 'taxa_5%', 'Total': 'total', 'Date': 'dia', 
            'Time': 'hora', 'Payment': 'forma_pagamento',
       'gross margin percentage': 'margem_lucro', 'gross income': 'renda_bruta', 'Rating': 'avaliacao'}
```

```python
dados = dados.rename(columns = renomear)
dados.head()
```

```python
dados.genero.unique()
```

```python
genero = {'Female': 'Feminino', 'Male': 'Masculino'}
```

```python
dados.genero = dados.genero.map(genero)
```

```python
dados.forma_pagamento.unique()
```

```python
forma_pagamento = {'Ewallet': 'Carteira Digital', 'Cash': 'Dinheiro', 'Credit card': 'Cartão de Crédito'}
```

```python
dados.forma_pagamento = dados.forma_pagamento.map(forma_pagamento)
```

```python
dados.tipo_produto.unique()
```

```python
tipo_produto = {'Health and beauty': 'Saúde e Beleza', 'Electronic accessories': 'Eletrônicos',
       'Home and lifestyle': 'Casa', 'Sports and travel': 'Esportes', 'Food and beverages': 'Alimentos',
       'Fashion accessories': 'Vestuário'}
```

```python
dados.tipo_produto = dados.tipo_produto.map(tipo_produto)
```

```python
dados.to_csv('vendas_supermercado.csv', index = False)
```
